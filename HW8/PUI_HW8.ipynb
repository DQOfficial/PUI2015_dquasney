```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
import numpy as np
plt.style.use('ggplot')
%pylab inline
```

```python
cafe = pd.read_csv('cafe.csv')
```

```python
fig = plt.figure()
```

```python
# change names of headings

old_names = ['pass_car_actual','pass_car_standard','light_truck_actual','light_truck_standard']
new_names = ['Passenger Cars - Actual', 'Passenger Cars - Required','Light Trucks - Actual', 'Light Trucks - Required']

cafe.rename(columns=dict(zip(old_names, new_names)), inplace=True)

x = cafe['year']
y = cafe[['Passenger Cars - Actual', 'Passenger Cars - Required','Light Trucks - Actual', 'Light Trucks - Required']]
k = np.ones(len(x))
```

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111)
ax.grid()

for z, i in enumerate(y.columns):
    ax.plot(x,y[i],alpha=0.8,linewidth=3,label=i);

for b in np.arange(10,36,5):    
    plt.plot(range(1985, 2013), [b] * len(range(1985, 2013)), "--", lw=0.5, color="black", alpha=0.3)
    
plt.ylabel('Average Fleet Miles per Gallon (MPG)', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.yticks(fontsize=14)
plt.xticks(fontsize=15)
plt.title('Corporate Average Fleet Emissions (CAFE) Levels, 1985-2013', fontsize = 20)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figtext(.07,.05,'Source: Environmental Protection Agency')
   
```
